```python
#Importación Librerias y archivo original
#Importante: Se utilizo el mismo archivo de los ejercicios pasados ya que se queria verificar la cantidad de palabras y ver el
#            nivel de complejidad que tenian los anteriores modelos al clasificar.

import pandas as pd
import numpy as np

from wordcloud import WordCloud, ImageColorGenerator

from nltk.corpus import stopwords
stopwords = stopwords.words("spanish")

import matplotlib.pyplot as plt
import PIL.Image
from IPython.display import display

import re 

data = pd.read_csv("C:/Users/johnv/reviews_vidjew_es.csv")
data.head()
```

```python
#Generación de Nube sobre conjunto inicial

palabras = data.review_body.str.cat(sep=" ")
wordcloud = WordCloud().generate(palabras)

plt.imshow(wordcloud)
plt.show()
```

```python
#Limpieza de texto

def pre_procesado(texto):
    texto = texto.lower()
    texto = re.sub(r"[\W\d_]+", " ", texto)
    texto = [palabra for palabra in texto.split() if palabra not in stopwords]
    texto = " ".join(texto)
    return texto

data['pp'] = data.review_body.apply(lambda texto: pre_procesado(texto))

data.head()
```

```python
#Generación de Nube con Texto limpio

palabras = data.pp.str.cat(sep=" ")
wordcloud = WordCloud(width=800, height=400,
                      max_font_size=120, max_words=150,
                      background_color="white", colormap="gist_heat",
                      stopwords=stopwords).generate(palabras) 

plt.figure(figsize=(10,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
